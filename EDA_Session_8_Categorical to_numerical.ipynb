**Label encoder**
- Label encoder package in **sickit learn**
- Sickit learn is a heart of ml
- Inside Sockit Learn we have method called **Preprocessing**
- Inder preprossing we have **LableEncoder**
- scikit-learn we will write as **sklearn**
- any sklearn package we have 3 steps
  - Step1: Read the package
  - Step2: Save the package
  - Step3: Apply Fit transform
    

```python
visa_df=pd.read_csv('Visadataset.csv')
```

```python
#step1: import the package
from sklearn.preprocessing import LabelEncoder

#Steo2: Save te package
le=LabelEncoder()

#Step3: Apply fit Transform
visa_df['case_status']=le.fit_transform(visa_df['case_status'])
```

**inverse-transform**
- In order to see inverse of labels we need to perform individually again

```python
visa_df=pd.read_csv('Visadataset.csv')
cols=visa_df.select_dtypes(include='object').columns
```

```python
for i in cols:
    from sklearn.preprocessing import LabelEncoder
    le=LabelEncoder()
    visa_df[i]=le.inverse_transform(visa_df[i])
```

```python
visa_df=pd.read_csv('Visadataset.csv')
cols=visa_df.select_dtypes(include='object').columns
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
visa_df['case_status']=le.fit_transform(visa_df['case_status'])
print(visa_df['case_status'])
print(le.inverse_transform(visa_df['case_status']))
```
